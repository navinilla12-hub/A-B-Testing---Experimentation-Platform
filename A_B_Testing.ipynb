```python
import pandas as pd
from scipy import stats
```

```python
# Load
df = pd.read_csv('/content/ab_data.csv')
```

```python
df.head()
```

```python
# Clean
df = df[df['group'] == df['landing_page'].apply(lambda x: 'treatment' if 'new' in x else 'control')]
```

```python
# Conversion rates
control = df[df['group'] == 'control']['converted']
treatment = df[df['group'] == 'treatment']['converted']
```

```python
print("Control Conversion:", control.mean())
print("Treatment Conversion:", treatment.mean())
```

```python
# T-test
t_stat, p_value = stats.ttest_ind(treatment, control)
```

```python
print("T-stat:", t_stat)
print("P-value:", p_value)
```

```python
if p_value < 0.05:
    print("Statistically Significant!")
else:
    print("Not Significant")
```

```python
lift = (treatment.mean() - control.mean()) / control.mean()
print("Lift:", lift)
```
